### Use Porter stemmer to perform stemming on each word

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
   
def porter_stemmer_func(file):
    ps = PorterStemmer()
    with open(file) as f:
        text = f.read()
        sent_text = nltk.sent_tokenize(text)
        for sentence in sent_text:
            tokenized_sent = nltk.word_tokenize(sentence)
            for word in tokenized_sent:
                print(word, " : ", ps.stem(word))
```

```python
porter_stemmer_func(quaid_file)
```

### Use Snowball stemmer to perform stemming on each word

```python
import nltk
from nltk.stem.snowball import SnowballStemmer
  
def snowball_stemmer_func(file):
    snow_stemmer = SnowballStemmer(language='english')
    with open(file) as f:
        text = f.read()
        sentence_text = nltk.sent_tokenize(text)
        for sentence in sentence_text:
            tokenized_sent = nltk.word_tokenize(sentence)
            for w in tokenized_sent:
                print(w, " : ", snow_stemmer.stem(w))
```

```python
snowball_stemmer_func(quaid_file)
```

### Write a program for spelling correction task using a look-up based approach



Reference: http://norvig.com/spell-correct.html

```python
import re
from collections import Counter

def words(text): return re.findall(r'\w+', text.lower())

WORDS = Counter(words(open('Data/big.txt').read()))

def P(word, N=sum(WORDS.values())): 
    return WORDS[word] / N

def correction(word): 
    return max(candidates(word), key=P)

def candidates(word): 
    return (known([word]) or known(edits1(word)) or known(edits2(word)) or [word])

def known(words): 
    return set(w for w in words if w in WORDS)

def edits1(word):
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)

def edits2(word): 
    "All edits that are two edits away from `word`."
    return (e2 for e1 in edits1(word) for e2 in edits1(e1))
```

```python
correction('speling')
```
